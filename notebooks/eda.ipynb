```python
from pathlib import Path
import polars as pl
import numpy as np

PROJECT_ROOT = Path().resolve().parent  # since you're in notebooks/
DATA_PATH = PROJECT_ROOT / "data" / "WA_Fn-UseC_-Telco-Customer-Churn.csv"

df = pl.read_csv(DATA_PATH)

print(df.shape)
df.head()
```

```python
df.describe()
```

```python
df.schema
```

```python
missing_df = df.select(
    pl.all().null_count().name.suffix("_nulls")
)
missing_df
```

```python
missing_pct = df.select([
    (pl.all().null_count() / df.height).name.suffix("_pct")
])
missing_pct
```

```python
target_dist = df.group_by("Churn").len()
target_dist
```

```python
target_dist = target_dist.with_columns(
    (pl.col("len") / pl.col("len").sum()).alias("ratio")
)
target_dist
```

```python
categorical_cols = [
    col for col, dtype in df.schema.items()
    if dtype == pl.Utf8
]

numerical_cols = [
    col for col, dtype in df.schema.items()
    if dtype in (pl.Int64, pl.Float64)
]
print(categorical_cols)
print(numerical_cols)
```

```python
cardinality = df.select([
    pl.col(col).n_unique().alias(col)
    for col in categorical_cols
])
cardinality
```

```python
print(df.select(numerical_cols).describe())
```

```python
corr = df.select(numerical_cols).corr()
print(corr)
```

```python
contract_churn = df.group_by(["Contract", "Churn"]).len()
contract_churn
```

```python
contract_churn = contract_churn.with_columns(
    (pl.col("len") / pl.col("len").sum().over("Contract")).alias("ratio")
)
contract_churn
```

```python
df.select([
    pl.col(col).n_unique().alias(col)
    for col in df.columns
]).sort(by=df.columns[0])
```

```python
df.select([
    pl.col(col).quantile(0.01).alias(f"{col}_p1")
    for col in numerical_cols
] + [
    pl.col(col).quantile(0.99).alias(f"{col}_p99")
    for col in numerical_cols
])
```
